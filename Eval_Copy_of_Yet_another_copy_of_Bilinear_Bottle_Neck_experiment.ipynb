```python
from omegaconf import OmegaConf
pre_training_config_path = "/content/deep-attentive-time-warping/code/conf/pre_training.yaml"
metric_learning_config_path = "/content/deep-attentive-time-warping/code/conf/metric_learning.yaml"
def change_batch_size(size: int,mode: int):
  if mode==0:
    path = pre_training_config_path
  else:
    path = metric_learning_config_path
  conf = OmegaConf.load(path)
  conf.batch_size = size
  # dumps to file
  with open(path, "w") as f:
    OmegaConf.save(conf, f)
```

```python
# imports
from pathlib import Path
import sys

# Get my_package directory path from Notebook
parent_dir = str(Path().resolve().parents[1])

# Add to sys.path
sys.path.insert(0, parent_dir)
```

```python
#import things
from deep_attentive_time_warping.code.model.proposed_unet_bottle_neck import ProposedModel
```

```python
# Load the model
model_path = "/content/deep_attentive_time_warping/result/001_Adiac/pre_training/Adiac_2024.01.25.22.57.25_lr_0.0001_ProposedModel_epoch_10_loss_0.0010.pkl"

#
model = ProposedModel(input_ch=1)
```

```python
import torch
model.load_state_dict(torch.load(model_path, map_location='cpu'))
```

```python
model
```

```python
# Freeze everything
for param in model.parameters():
    param.requires_grad = False
```

```python
model.unet.projector
```

```python
# Unfreeze weight and bias of the projector
for param in model.unet.projector.parameters():
    param.requires_grad = True
```

```python
for name, param in model.named_parameters():
  print("{} {}".format(name, param.requires_grad))
```
